```python
import pandas as pd
import shutil
import os
```

```python
SOURCE_DATA_PATH = '../datasets/raw/PAD-UFES-20/'
DESTINATION_DATASET_PATH = '../datasets/processed/PAD-UFES-20/'
```

```python
if not os.path.exists(DESTINATION_DATASET_PATH):

    # if the demo_folder directory is not present
    # then create it
    os.makedirs(DESTINATION_DATASET_PATH, exist_ok=True)

if not os.path.exists(DESTINATION_DATASET_PATH + 'melanoma'):

    # if the demo_folder directory is not present
    # then create it
    os.makedirs(DESTINATION_DATASET_PATH + 'melanoma', exist_ok=True)

if not os.path.exists(DESTINATION_DATASET_PATH + 'benign'):

    # if the demo_folder directory is not present
    # then create it
    os.makedirs(DESTINATION_DATASET_PATH + 'benign', exist_ok=True)
```

EDA

```python
df = pd.read_csv(SOURCE_DATA_PATH + 'metadata.csv')
df.head()
```

```python
df.columns
```

```python
df.region.unique()
```

```python
df.diagnostic.unique()
```

```python
# filtering data in acral regions
hand_foot = df[df.region.isin(['HAND', 'FOOT'])]
hand_foot
```

```python
# removing other malignant skin lesions besides Melanoma
hand_foot2 = hand_foot[~hand_foot.diagnostic.isin(['BCC', 'SCC'])]
hand_foot2
```

```python
hand_foot2.diagnostic.value_counts()
```

Moving filtered data

```python
benign_imgs = hand_foot2[hand_foot2.diagnostic != 'MEL'].img_id.values
mel_imgs = hand_foot2[hand_foot2.diagnostic == 'MEL'].img_id.values
```

```python
for img_path in benign_imgs:
    shutil.copy(SOURCE_DATA_PATH + 'images/'+img_path, DESTINATION_DATASET_PATH + 'benign/')

for img_path in mel_imgs:
    shutil.copy(SOURCE_DATA_PATH + 'images/'+img_path, DESTINATION_DATASET_PATH + 'melanoma/')
```

```python
hand_foot2
```

```python
hand_foot2['is_dermoscopic'] = False
hand_foot2['label'] = hand_foot2.diagnostic.apply(lambda x: 1 if x == 'MEL' else 0)
hand_foot2['lesion_filename'] = hand_foot2.img_id.apply(lambda x: DESTINATION_DATASET_PATH[3:] + '{}/'.format('melanoma' if x in mel_imgs else 'benign') + x)
```

```python
hand_foot2
```

```python
# save metadata
hand_foot2.to_csv(DESTINATION_DATASET_PATH + 'metadata.csv', index=False)
```
